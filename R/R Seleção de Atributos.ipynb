```r
library('e1071')
#library('FSelector')
```

```r
credito = read.csv('Credit.csv', sep=',', header = T)
```

```r
head(credito)
```

```r
amostra = sample(2, 1000, replace=T, prob=c(0.7, 0.3))
credito_treino = credito[amostra==1, ]
credito_teste = credito[amostra==2, ]
modelo = svm(class ~ ., credito_treino)
```

```r
modelo
```

```r
modelo
```

```r
print(modelo)
```

```r
predição = predict(modelo, newdata= credito_teste)
```

```r
predição
```

```r
confusao  = table(credito_teste$class, predição)
confusao
```

```r
taxa_acerto = (confusao[1] + confusao[4])/ sum(confusao)
```

```r
taxa_acerto
```

```r
taxa_erro = (confusao[2] + confusao[3])/sum(confusao)
```

```r
taxa_erro
```
